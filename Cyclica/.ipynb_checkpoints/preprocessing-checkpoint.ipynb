### deal with imbalanced data

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
```

```python
from sklearn import metrics
from sklearn.metrics import make_scorer
```

```python
def make_auc_pr(y_true, y_pred):
    precision, recall, _ = metrics.precision_recall_curve(y_true, y_pred)
    return metrics.auc(recall, precision)
auc_pr = make_scorer(make_auc_pr)
```

```python
rf = RandomForestClassifier()
```

```python
def fit_model(over_strategy, under_strategy, rf):
    over = SMOTE(sampling_strategy=over_strategy)
    under = RandomUnderSampler(sampling_strategy=under_strategy)

    pipeline = Pipeline(steps=[['o', over], ['u', under]])
    X_train, y_train = pipeline.fit_resample(scaled_X, y)

    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=1, random_state=20902490)
    scores = cross_validate(rf, X_train, y_train, scoring={'roc_auc': 'roc_auc',
                                                            'auc_pr': auc_pr}, cv=cv, n_jobs=-1)
    return scores
```

```python
scores = fit_model(0.1, 0.5, rf)
print(scores)
```

```python
def find_mean(scores):
    print("test_roc_auc: ", scores['test_roc_auc'].mean())
    print("test_auc_pr: ", scores['test_auc_pr'].mean())    
```

```python
find_mean(scores)
```

### tune the model

```python
# find the best sampling strategy
over_strategies = [0.1, 0.2, 0.3, 0.5, 0.6, 0.8]
under_strategies = [0.3, 0.5, 0.6, 0.8, 0.9, 0.7]
```

```python
for i in range(6):
    scores = fit_model(over_strategies[i], under_strategies[i], rf)
    print("smote sampling strategy: ", over_strategies[i], 
          " under sampling strategy: ", under_strategies[i])
    find_mean(scores)
```

```python
for i in [4, 5]:
    scores = fit_model(over_strategies[i], under_strategies[i], rf)
    print("smote sampling strategy: ", over_strategies[i], 
          " under sampling strategy: ", under_strategies[i])
    find_mean(scores)
```

```python
scores = fit_model(0.8, 0.9, rf)
find_mean(scores)
```
